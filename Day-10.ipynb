```python
class Solution(object):
    def isAnagram(self, s, t):
        if sorted(s) == sorted(t):
            return True
        else:
            return False
```

```python
sol = Solution()
print(sol.isAnagram("anagram", "nagaram"))  
print(sol.isAnagram("rat", "car"))          
```
