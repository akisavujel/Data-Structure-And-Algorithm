  First Unique Character in String
  

  <span style="display: block; width: 70%; height: 1px; background: linear-gradient(90deg, transparent, #ffb8d9, transparent); margin: 12px auto;"></span>

  <span style="font-size: 13px; display: block; font-weight: 400; color: #f093b0; letter-spacing: 3px;">(=^･ω･^=) ✦ (=^･ω･^=)</span>

</h1>

```python
class Solution(object):
    def firstUniqChar(self, s):
        for i, j in enumerate(s): # I is the index value and j means the unique characters
            if s.count(j) == 1:   # If count of character is one 
                return i          # Return index of character  
        return -1                 # If no unique charater exist
```

```python
s = "loveleetcode"
sol = Solution()
print(sol.firstUniqChar(s)) 
```
